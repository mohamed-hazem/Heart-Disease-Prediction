# Import Modules

```python
import pandas as pd
import pycaret.classification as pyc
```

# Load Data

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

# Train and Evaluate Model

```python
cat_features = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
```

```python
clf = pyc.setup(df, target='HeartDisease',
             categorical_features=cat_features,
             train_size=0.9)
```

```python
best_model = pyc.compare_models()
```

# Test Model

```python
pyc.predict_model(best_model, df.tail())
```
